```python
def categorize(age):
    if age < 18.0:
        return 'young'
    elif age < 34.0:
        return 'adult'
    elif age < 54.0:
        return 'middle-aged'
    else:
        return 'senior'
    
```

```python
df['age_group']=df['age'].apply(categorize)
```
